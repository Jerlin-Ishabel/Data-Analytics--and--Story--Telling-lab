```python
#Scatter Plots:scatter()
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,1,2,3,4,5])
y=np.array([0,8,12,20,26,38])
plt.scatter(x,y)
plt.show()
```

```python
#Compare Plots

import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,1,2,3,4,5])
y=np.array([0,2,8,1,14,7])
plt.scatter(x,y,color='violet')

x=np.array([12,6,8,11,8,3])
y=np.array([5,6,3,7,17,19])
plt.scatter(x,y,color='green')
plt.show()
```

```python
#Color Each Dots

x=np.array([0,1,2,3,4,5])
y=np.array([0,2,8,1,14,7])
mycolor=['red','green','blue','lime','aqua','yellow']
plt.scatter(x,y,color=mycolor)
plt.show()
```

```python
#Size(s)

x=np.array([0,1,2,3,4,5])
y=np.array([0,2,8,1,14,7])
mycolor=['red','green','blue','lime','aqua','yellow']
size=[10,60,120,80,20,190]
plt.scatter(x,y,color=mycolor,s=size)
plt.show()
```

```python
#Alpha

x=np.array([0,1,2,3,4,5])
y=np.array([0,2,8,1,14,7])
mycolor=['red','green','blue','lime','aqua','yellow']
size=[10,60,120,80,20,190]
plt.scatter(x,y,color=mycolor,s=size,alpha=0.3)
plt.show()
```

```python
#Create Bar

import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y)
plt.show()
```

```python
#Create Horizontal Bar

import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.barh(x,y)
plt.show()
```

```python
#Bar Width

import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y,width=0.1)
plt.show()
```

```python
#Histogram:hist()

import matplotlib.pyplot as plt
import numpy as np
x=np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```

```python
#Creating Pie Charts:pie()

import matplotlib.pyplot as plt
import numpy as np
x=np.array([35,25,25,15])
plt.pie(x)
plt.show()
```

```python
#labels

import matplotlib.pyplot as plt
import numpy as np
x=np.array([35,25,25,15])
mylabels=["Apples","Bananas","Cherries","Dates"]
plt.pie(x,labels=mylabels)
plt.show()
```

```python
#Start Angle

import matplotlib.pyplot as plt
import numpy as np
x=np.array([35,25,25,15])
mylabels=["Apples","Bananas","Cherries","Dates"]
plt.pie(x,labels=mylabels,startangle=90)
plt.show()
```

```python
#Explode

import matplotlib.pyplot as plt
import numpy as np
x=np.array([35,25,25,15])
mylabels=["Apples","Bananas","Cherries","Dates"]
myexplode=[0.2,0,0,0]
plt.pie(x,labels=mylabels,startangle=90,explode=myexplode)
plt.show()
```

```python
#Shadow

import matplotlib.pyplot as plt
import numpy as np
x=np.array([35,25,25,15])
mylabels=["Apples","Bananas","Cherries","Dates"]
myexplode=[0.2,0,0,0]
plt.pie(x,labels=mylabels,startangle=90,explode=myexplode,shadow=True)
plt.show()
```

```python
#Colors

import matplotlib.pyplot as plt
import numpy as np
x=np.array([35,25,25,15])
mylabels=["Apples","Bananas","Cherries","Dates"]
mycolors=["black","hotpink","b","#4CAF50"]
myexplode=[0.2,0,0,0]
plt.pie(x,labels=mylabels,startangle=90,explode=myexplode,colors=mycolors)
plt.show()
```

```python
#legend

import matplotlib.pyplot as plt
import numpy as np
x=np.array([35,25,25,15])
mylabels=["Apples","Bananas","Cherries","Dates"]
plt.pie(x,labels=mylabels)
plt.legend()
plt.show()
```

```python
#Legend With Header

import matplotlib.pyplot as plt
import numpy as np
x=np.array([35,25,25,15])
mylabels=["Apples","Bananas","Cherries","Dates"]
plt.pie(x,labels=mylabels)
plt.legend(title="Four Fruits:")
plt.show()
```
