# Week 5: Data Visualization Using Numpy

# 1: Numpy Histogram

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(1000)
plt.hist(data, bins=30, alpha=0.5, color='blue')
plt.title("Histogram from Numpy Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

# 2: Numpy Scatter Plot

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)
plt.title("Scatter Plot from Numpy Data")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# 3: Numpy Line Plot

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title("Line Plot from Numpy Data")
plt.xlabel("X-axis")
plt.ylabel("Sine Values")
plt.show()
```

# Week 6: Data Visualization Using Pandas

# 1: Bar Plot with Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame({
'Category': ['A', 'B', 'C', 'D'],

'Values': [4, 7, 1, 8]
})
df.plot(kind='bar', x='Category', y='Values', color='purple')
plt.title("Bar Plot from Pandas DataFrame")
plt.show()
```

# 2: Line Plot with Pandas

```python
df = pd.DataFrame({
'X': [1, 2, 3, 4],
'Y': [10, 20, 25, 30]
})
df.plot(kind='line', x='X', y='Y', marker='o')
plt.title("Line Plot from Pandas DataFrame")
plt.show()
```

# 3: Pie Chart with Pandas

```python
df = pd.DataFrame({
'Categories': ['A', 'B', 'C', 'D'],
'Values': [20, 30, 25, 25]
})
df.plot.pie(y='Values', labels=df['Categories'], autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart from Pandas DataFrame")
plt.ylabel('')
plt.show()
```
