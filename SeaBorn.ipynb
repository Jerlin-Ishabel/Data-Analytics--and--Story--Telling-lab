# 1.UniVariate
HISTOGRAM

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
data=np.random.randn(1000)
sns.histplot(data,bins=30,color="green")
plt.title("univariate plot-Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

# KDE PLOT(Kernel Density Estimate)

```python
sns.kdeplot(data,shade=True,color="pink")
plt.title("univariate plot-KDE")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

# BOX PLOT

```python
sns.boxplot(x=data,color="orange")
plt.title("univariate plot-Box Plot")
plt.xlabel("Value")
plt.show()
```

# VIOLIN PLOT

```python
sns.violinplot(x=data,color="gold")
plt.title("univariate plot-Violin Plot")
plt.xlabel("Value")
plt.show()
```

# 2.BiVariate
SCATTER PLOT

```python
x=np.random.rand(100)
y=np.random.rand(100)
sns.scatterplot(x=x,y=y,color="violet")
plt.title("Bivariate plot- Scatter plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# LINE PLOT

```python
x=np.linspace(0,10,100)
y=np.sin(x)
sns.lineplot(x=x,y=y,color="blue")
plt.title("Bivariate plot- Line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# HEAT MAP

```python
data=np.random.rand(10,10)
sns.heatmap(data,annot=True,cmap="coolwarm")
plt.title("Bivariate plot- Heat Map")
plt.show()
```

# BAR PLOT

```python
categories=["A","B","C"]
values=[10,20,15]
sns.barplot(x=categories,y=values,palette="muted")
plt.title("Bivariate plot- Bar Plot")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

# PAIR PLOT:(Special Bivariate Plot)

```python
import pandas as pd
data=pd.DataFrame({"x":np.random.rand(50),"y":np.random.rand(50),"z":np.random.rand(50)})
sns.pairplot(data)
plt.suptitle("Bivariate plot- Pair Plot")
plt.show()
```
