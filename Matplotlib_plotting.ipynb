```python
#1.Plotting Line
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,6])
y=np.array([0,250])
plt.plot(x,y)
plt.show()
```

```python
#2.plotting without line
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,6])
y=np.array([0,250])
plt.plot(x,y,'o')
plt.show()
```

```python
#3.Multiple Points
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,6,8])
y=np.array([3,8,1,10])
plt.plot(x,y)
plt.show()
```

```python
#4.Default x-Axis
y=np.array([3,8,1,10,5,7])
plt.plot(y)
plt.show()
```

```python
#5.Matplotlib Marker
y=np.array([3,8,1,10])
plt.plot(y,marker='o')
plt.show()
```

```python
#6.Format String Fmt
y=np.array([3,8,1,10])
plt.plot(y,'o:r')
plt.show()
```

```python
#7.Marker Size(ms)
y=np.array([3,8,1,10])
plt.plot(y,marker='o',ms=20)
plt.show()
```

```python
#8.Marker Edge Color(mec)
y=np.array([3,8,1,10])
plt.plot(y,marker='o',ms=20,mec='r')
plt.show()
```

```python
#9.Marker Face Color(mfc)
y=np.array([3,8,1,10])
plt.plot(y,marker='o',ms=20,mfc='r')
plt.show()
```

```python
#10.Linestyle Argument(ls)
y=np.array([3,8,1,10])
plt.plot(y,ls='dotted')
plt.show()
```

```python
#11.Line color Argument(c)
y=np.array([3,8,1,10])
plt.plot(y,c='r')
plt.show()
```

```python
#12.Line Width Argument(lw)
y=np.array([3,8,1,10])
plt.plot(y,lw='20.6')
plt.show()
```

```python
#13.Multiple Width Argument
x1=np.array([0,1,2,3])
y1=np.array([3,8,1,10])
x2=np.array([0,1,2,3])
y2=np.array([6,2,7,11])
plt.plot(x1,y1,color='r')
plt.plot(x2,y2,color='g')
plt.show()
```

```python
#14.create labels
x=np.array([0,1,2,3,4,5])
y=np.array([0,8,12,20,26,38])
plt.plot(x,y)
plt.xlabel("overs")
plt.ylabel("runs")
plt.show()
```

```python
#15.Create Title- title()
x=np.array([0,1,2,3,4,5])
y=np.array([0,8,12,20,26,38])
plt.plot(x,y)
plt.xlabel("overs")
plt.ylabel("runs")
plt.title("sport Data")
plt.show()
```

```python
#16.set fornt properties for title and labels
x=np.array([0,1,2,3,4,5])
y=np.array([0,8,12,20,26,38])
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':15}
plt.plot(x,y)
plt.xlabel("overs",fontdict=font2)
plt.ylabel("runs",fontdict=font2)
plt.title("sport Data",fontdict=font1)
plt.show()
```

```python
#17.position the title
x=np.array([0,1,2,3,4,5])
y=np.array([0,8,12,20,26,38])
plt.plot(x,y)
plt.xlabel("overs")
plt.ylabel("runs")
plt.title("sport Data",loc='left')
plt.show()
```

```python
#18.add grid lines ta a plot
x=np.array([0,1,2,3,4,5])
y=np.array([0,8,12,20,26,38])
plt.plot(x,y)
plt.xlabel("overs")
plt.ylabel("runs")
plt.title("sport Data",loc='left')
plt.grid()
plt.show()
```

```python
#19.display multiple plots
#plot1
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x,y)
#plot2
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

```python
#20.super title
#plot1
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x,y)
#plot2
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.suptitle("my data")
plt.show()
```
